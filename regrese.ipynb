```python
F=[lambda x:1, lambda x:x, lambda x:x*x]  #Definuji svůj seznam funkcí
k=len(F)

xx=vector([-1,0,1,2])  #Naměřené hodnoty
y=vector([3,1,-1,1])
n=len(xx)
```

```python
A=matrix([[F[j](xx[i]) for j in range(k)] for i in range (n)])
A
```

```python
#Chceme řešit rovnici Ap=y, kde neznámou je vektor p, ovšem ta nemá řešení
#Místo toho řešíme rovnici Bp=z, kde
B=A.transpose()*A
z=A.transpose()*y
```

```python
p=B.solve_right(z)
p
```

```python
def f(x):
    return(sum([p[i]*F[i](x) for i in range(k)]))
plot(f(x),(x,xx[0],xx[n-1]))+list_plot(list(zip(xx,y)),color='red')
```

```python
sum([(y[i]-f(xx[i]))^2 for i in range(n)]) #chyba ||Ap-y||^2
```
